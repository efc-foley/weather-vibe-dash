```python
# import pandas library
import pandas as pd

# read the csv file
online_orders = pd.read_csv("online_orders.csv")
online_orders.head(10)
```

```python
# one of the method to analyze dataframe is to make them into a list/ lists;
# I will make each row as a list for this demo purpose

order_lists = online_orders.values.tolist()
```

```python
# Human coded solution

# assign a starting number 0 to total
total = 0

# for loop is used for iterating over a sequence
# this for loop let row value to iterating over a list
for row in order_lists:
    
    # index 2 refers to the value in the city column in each row
    # remember python index starts at 0
    city = row[2]
    
    # since some csv has blank spaces in its values, is it safer to use 'New York' in city than city == 'New York' 
    if 'New York' in city:
        # index 4 refers to the value in the price column in each row
        price = float(row[4])
        
        total = total + price
        
        print('Order amount: $', row[4],'running total: $', round((total), 2))
        
print('\nTotal price of online orders in New York is $', total)
```
